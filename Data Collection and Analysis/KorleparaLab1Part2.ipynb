```r
data<-read.csv("Genetic04.csv")

slices <- c(18068,1896,348,228,2292,921)
colors <- c("red", "orangered", "yellow","limegreen"," chartreuse ","green")
s<- sum(slices)
perc<- round((slices/s)*100)
pie(slices, col=colors, main="Influenza Positive Specimens Reported by U.S. Health Laboratories, Cumulative, 2017-18 Season")

subdata<-subset(data,data$X.Sub.type=="B/Victoria")
slices<-c(as.numeric(sub("%","",subdata$X..of..Sub.type.Total)))
lb<-subdata$Genetic_Group
number<-subdata$Number
colors <- c("Violet", "Blue")
s<- sum(slices)
perc<- round((slices/s)*100)
label <- paste(number, sep="")
lb <- paste(lb, "\n", perc)
lb <- paste(lb, "%", sep="")
lb <- paste(lb, "\n", label)
pie(slices, labels = lb, col=colors, main="Influenza B Victoria")


subdata<-subset(data,data$X.Sub.type=="B/Yamagata")
slices<-as.numeric(sub("%","",subdata$X..of..Sub.type.Total))
slices<-slices[!is.na(slices)]
lb<-subdata$Genetic_Group
number<-subdata$Number
colors <- c("cyan")
s<- sum(slices)
perc<- round((slices/s)*100)
label <- paste(number, sep="")
lb <- paste(lb, "\n", perc)
lb <- paste(lb, "%", sep="")
lb <- paste(lb, "\n", label)
pie(slices, labels = lb, col=colors, main="Influenza B Yamagata")


subdata<-subset(data,data$X.Sub.type=="H1pdm09")
slices<-as.numeric(sub("%","",subdata$X..of..Sub.type.Total))
slices<-slices[!is.na(slices)]
lb<-subdata$Genetic_Group
number<-subdata$Number
colors <- c("pink")
s<- sum(slices)
perc<- round((slices/s)*100)
label <- paste(number, sep="")
lb <- paste(lb, "\n", perc)
lb <- paste(lb, "%", sep="")
lb <- paste(lb, "\n", label)
pie(slices, labels = lb, col=colors, main="Influenza A (H1N1)")


subdata<-subset(data,data$X.Sub.type=="H3")
slices<-as.numeric(sub("%","",subdata$X..of..Sub.type.Total))
slices<-slices[!is.na(slices)]
lb<-subdata$Genetic_Group
number<-subdata$Number
colors <- c("blue", "red", "green")
s<- sum(slices)
perc<- round((slices/s)*100)
label <- paste(number, sep="")
lb <- paste(lb, "\n", perc)
lb <- paste(lb, "%", sep="")
lb <- paste(lb, "\n", label)
pie(slices, labels=lb, col=colors, main="Influenza A (H3N2)")
```

```r
install.packages('maps', repos = "http://cran.us.r-project.org")
install.packages('mapdata', repos = "http://cran.us.r-project.org")
install.packages('ggmap', repos = "http://cran.us.r-project.org")
library(maps)
library(mapdata)
library(ggmap)
us <- map_data("state")
data<-read.csv("heatmap.csv")
data<-data[!(data$STATENAME=="Alaska"),]
data<-data[!(data$STATENAME=="Hawaii"),]
data<-data[!(data$STATENAME=="Virgin Islands"),]
data<-data[!(data$STATENAME=="Puerto Rico"),]
ILIActivityLevel=data$ACTIVITY.LEVEL

data[("region")] =tolower(data$STATENAME)
arr=data
gg <- ggplot()
gg <- gg + geom_map(data=us, map=us, aes(x=long, y=lat, map_id=region), fill="black", color="black", size=0.15)
gg <- gg + geom_map(data=arr, map=us, aes(fill=ILIActivityLevel, map_id=region), color="black", size=0.15)

gg <- gg + labs(x=NULL, y=NULL)
gg <- gg + coord_map("albers", lat0 = 39, lat1 = 45) 
gg <- gg + theme(panel.border = element_blank())
gg <- gg + theme(panel.background = element_blank())
gg <- gg + theme(axis.ticks = element_blank())
gg <- gg + theme(axis.text = element_blank())
gg
```
